```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('SocialMedia.csv')
```

```python
df.describe()
```

```python
df.drop(columns=['permalink clicks','app opens','app installs','follows','email tweet','dial phone'],inplace=True)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
sns.histplot(df['impressions'],kde=True)
```

```python
df['impressions'].fillna(df['impressions'].median(),inplace=True)
```

```python
df['engagements'].fillna(df['engagements'].median(), inplace=True)
df['engagement rate'].fillna(df['engagement rate'].median(), inplace=True)
df['retweets'].fillna(df['retweets'].median(), inplace=True)
df['media views'].fillna(df['media views'].median(), inplace=True)
df['media engagements'].fillna(df['media engagements'].median(), inplace=True)
```

```python
df["date"]=df["time"].str.split(" ").str[0]
df["time_extracted"]=df["time"].str.split(" ").str[1]
```

```python
df
```

```python
df["Day"]=df["date"].str.split("-").str[2]
df["Month"]=df["date"].str.split("-").str[1]
df["Year"]=df["date"].str.split("-").str[0]
```

```python
df
```

```python
invalid_times = df[~df['time'].str.match(r'\d{4}-\d{2}-\d{2} \d{2}:\d{2} \+\d{4}')]
print(invalid_times[['time']])
```

```python

df['time'] = pd.to_datetime(df['time'], utc=True, errors='coerce')

df = df.dropna(subset=['time'])
```

```python
df['date'] = df['time'].dt.date.astype(str)
df['time_extracted'] = df['time'].dt.strftime('%H:%M')
df['Day'] = df['time'].dt.day
df['Month'] = df['time'].dt.month
df['Year'] = df['time'].dt.year
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.to_csv('cleaned_socialmedia2.csv')
```
